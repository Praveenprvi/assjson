```python
[
    {
        "Name": "John Doe",
        "DOB": "1990-01-01",
        "Height": 170,
        "City": "New York",
        "State": "New York"
    },
    {
        "Name": "Jane Smith",
        "DOB": "1992-05-12",
        "Height": 165,
        "City": "Los Angeles",
        "State": "California"
    },
    {
        "Name": "Michael Johnson",
        "DOB": "1985-09-30",
        "Height": 180,
        "City": "Chicago",
        "State": "Illinois"
    },
    {
        "Name": "Emily Davis",
        "DOB": "1993-07-19",
        "Height": 160,
        "City": "Houston",
        "State": "Texas"
    },
    {
        "Name": "David Wilson",
        "DOB": "1991-03-25",
        "Height": 175,
        "City": "Philadelphia",
        "State": "Pennsylvania"
    }
]
import json

class Employee:
    def __init__(self, name, dob, height, city, state):
        self.name = name
        self.dob = dob
        self.height = height
        self.city = city
        self.state = state

# Read employee data from the JSON file
with open('employee.json') as file:
    data = json.load(file)

# Create a list to store Employee objects
employees = []

# Iterate over each employee data and create Employee objects
for employee_data in data:
    employee = Employee(employee_data['Name'], employee_data['DOB'], employee_data['Height'], employee_data['City'], employee_data['State'])
    employees.append(employee)

# Print the list of Employee objects
for employee in employees:
    print(f"Name: {employee.name}")
    print(f"DOB: {employee.dob}")
    print(f"Height: {employee.height}")
    print(f"City: {employee.city}")
    print(f"State: {employee.state}")
    print()

```
